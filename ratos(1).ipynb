```python
import os
import glob
import pandas as pd
import numpy as np

# =========
# 1. Carregar anota√ß√µes
# =========
ann_files = glob.glob("./MABe-mouse-behavior-detection/train_annotation/*/*.parquet")

annotations = []
for f in ann_files:
    df = pd.read_parquet(f)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    annotations.append(df)

annotations = pd.concat(annotations, ignore_index=True)
print("Annotations:", annotations.shape)
print(annotations.head())

# =========
# 2. Carregar tracking
# =========
trk_files = glob.glob("./MABe-mouse-behavior-detection/train_tracking/*/*.parquet")[:5]

trackings = []
for f in trk_files:
    df = pd.read_parquet(f)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    trackings.append(df)

trackings = pd.concat(trackings, ignore_index=True)

# üîß Renomear video_frame ‚Üí frame
trackings = trackings.rename(columns={"video_frame": "frame"})

print("Tracking:", trackings.shape)
print(trackings.head())

# =========
# 3. Expandir labels frame-a-frame
# =========
expanded = []
for _, row in annotations.iterrows():
    frames = list(range(row.start_frame, row.stop_frame + 1))
    expanded.append(pd.DataFrame({
        "frame": frames,
        "agent_id": row.agent_id,
        "target_id": row.target_id,
        "action": row.action,
        "source_file": row.source_file,
        "source_folder": row.source_folder
    }))
expanded = pd.concat(expanded, ignore_index=True)
print("Expanded labels:", expanded.shape)

# =========
# 4. Merge tracking + labels
# =========
merged = trackings.merge(
    expanded,
    on=["frame", "source_file", "source_folder"],
    how="left"
)
print("Merged:", merged.shape)

# =========
# 5. Multi-label binariza√ß√£o
# =========
frame_actions = merged.groupby(
    ["source_folder", "source_file", "frame"]
)["action"].apply(lambda x: list(set(x.dropna()))).reset_index()

all_actions = sorted({a for acts in frame_actions["action"] if isinstance(acts, list) for a in acts})
action_to_idx = {a: i for i, a in enumerate(all_actions)}

y = np.zeros((len(frame_actions), len(all_actions)), dtype=np.uint8)
for i, acts in enumerate(frame_actions["action"]):
    if isinstance(acts, list):
        for a in acts:
            if a in action_to_idx:  
                y[i, action_to_idx[a]] = 1

print("Classes:", all_actions)
print("y shape:", y.shape)
print("y exemplo:", y[:5])

# =========
# 6. Features X
# =========
X = merged.drop(columns=["action", "agent_id", "target_id"]).select_dtypes(include=[np.number])
print("X shape:", X.shape)
```

```python
import os
import glob
import pandas as pd
import numpy as np
from tqdm import tqdm

# =========
# 1. Carregar anota√ß√µes
# =========
ann_files = glob.glob("./MABe-mouse-behavior-detection/train_annotation/*/*.parquet")
annotations = []
for f in tqdm(ann_files, desc="Lendo anota√ß√µes"):
    df = pd.read_parquet(f)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    annotations.append(df)
annotations = pd.concat(annotations, ignore_index=True)

# Expandir labels frame a frame
expanded = []
for _, row in tqdm(annotations.iterrows(), total=len(annotations), desc="Expandindo labels"):
    frames = range(row.start_frame, row.stop_frame + 1)
    expanded.append(pd.DataFrame({
        "frame": frames,
        "action": row.action,
        "source_file": row.source_file,
        "source_folder": row.source_folder
    }))
expanded = pd.concat(expanded, ignore_index=True)

# =========
# 2. Definir classes √∫nicas (para y)
# =========
all_actions = sorted(expanded["action"].unique())
action_to_idx = {a: i for i, a in enumerate(all_actions)}

# =========
# 3. Fun√ß√£o para processar tracking de um arquivo por vez
# =========
def process_tracking_file(f):
    df = pd.read_parquet(f)
    df.rename(columns={"video_frame": "frame"}, inplace=True)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    
    # Pivot: um vetor por frame contendo todas as coordenadas
    pivot = df.pivot_table(
        index=["source_folder", "source_file", "frame"],
        columns=["mouse_id", "bodypart"],
        values=["x", "y"]
    )
    pivot.columns = [f"{c[0]}_{c[1]}_{c[2]}" for c in pivot.columns]  # flatten
    
    # Labels multi-hot
    acts = expanded[
        (expanded.source_file == os.path.basename(f)) &
        (expanded.source_folder == os.path.basename(os.path.dirname(f)))
    ]
    frame_actions = acts.groupby("frame")["action"].apply(list)
    
    y = np.zeros((len(pivot), len(all_actions)), dtype=np.uint8)
    for i, fr in enumerate(pivot.index.get_level_values("frame")):
        if fr in frame_actions.index:
            for a in frame_actions.loc[fr]:
                y[i, action_to_idx[a]] = 1
    
    return pivot.reset_index(), y

# =========
# 4. Processar todos os arquivos e salvar
# =========
X_list, y_list = [], []
trk_files = glob.glob("./MABe-mouse-behavior-detection/train_tracking/*/*.parquet")

for f in tqdm(trk_files, desc="Processando tracking"):
    try:
        Xf, yf = process_tracking_file(f)
        X_list.append(Xf)
        y_list.append(yf)
    except Exception as e:
        print(f"Erro em {f}: {e}")

# Concatenar tudo
X = pd.concat(X_list, ignore_index=True)
y = np.vstack(y_list)

print("Shape final X:", X.shape)
print("Shape final y:", y.shape)
print("Classes:", all_actions)

# =========
# 5. Salvar para n√£o refazer sempre
# =========
X.to_parquet("X_features.parquet")
np.save("y_labels.npy", y)
```
