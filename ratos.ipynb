```python
import glob
import pandas as pd
import os

# 1. Procurar todos os .parquet recursivamente
files = glob.glob("./MABe-mouse-behavior-detection(1)/train_annotation/**/*.parquet", recursive=True)

print("Qtd de arquivos encontrados:", len(files))
print("Exemplos:", files[:5])  # só para inspecionar

# 2. Carregar cada parquet
df_list = []
for f in files:
    try:
        df_tmp = pd.read_parquet(f)
        df_tmp["source_file"] = os.path.basename(f)   # de qual arquivo veio
        df_tmp["source_folder"] = os.path.basename(os.path.dirname(f))  # de qual pasta veio
        df_list.append(df_tmp)
    except Exception as e:
        print(f"Erro lendo {f}: {e}")

# 3. Concatenar em um único DataFrame
df = pd.concat(df_list, ignore_index=True)

print("Shape final:", df.shape)
print(df.head())
```

```python
import os
import glob
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer

# =========
# 1. Carregar anotações
# =========
ann_files = glob.glob("./MABe-mouse-behavior-detection(1)/train_annotation/*/*.parquet")

annotations = []
for f in ann_files:
    df = pd.read_parquet(f)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    annotations.append(df)

annotations = pd.concat(annotations, ignore_index=True)
print("Annotations:", annotations.shape)
print(annotations.head())

# =========
# 2. Carregar tracking
# =========
trk_files = glob.glob("./MABe-mouse-behavior-detection(1)/train_tracking/*/*.parquet")

trackings = []
for f in trk_files:
    df = pd.read_parquet(f)
    df["source_file"] = os.path.basename(f)
    df["source_folder"] = os.path.basename(os.path.dirname(f))
    trackings.append(df)

trackings = pd.concat(trackings, ignore_index=True)
print("Tracking:", trackings.shape)
print(trackings.head())

# =========
# 3. Expandir labels frame-a-frame
# =========
expanded = []
for _, row in annotations.iterrows():
    frames = list(range(row.start_frame, row.stop_frame + 1))
    expanded.append(pd.DataFrame({
        "frame": frames,
        "agent_id": row.agent_id,
        "target_id": row.target_id,
        "action": row.action,
        "source_file": row.source_file,
        "source_folder": row.source_folder
    }))
expanded = pd.concat(expanded, ignore_index=True)
print("Expanded labels:", expanded.shape)

# =========
# 4. Merge tracking + labels
# =========
merged = trackings.merge(
    expanded,
    left_on=["frame", "source_file", "source_folder"],
    right_on=["frame", "source_file", "source_folder"],
    how="left"
)

# =========
# 5. Multi-label binarização
# =========
# Agrupar por frame (caso várias ações no mesmo frame)
frame_actions = merged.groupby(["source_folder", "source_file", "frame"])["action"].apply(lambda x: list(set(x.dropna()))).reset_index()

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(frame_actions["action"])

print("Classes:", mlb.classes_)
print("y shape:", y.shape)

# =========
# 6. Features X
# =========
# Exemplo: pegar todas as coordenadas dos pontos do tracking como features
X = merged.drop(columns=["action", "agent_id", "target_id"]).select_dtypes(include=[np.number])

print("X shape:", X.shape)
```
